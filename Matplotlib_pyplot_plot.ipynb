載入需求套件

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

建立資料：  
* `np.linspace(start, stop, num)`，在start與stop之間產生num筆資料
* `np.sin`，即是將值利用sin函數計算

```python
x = np.linspace(0.01, 20, 100)
y = np.sin(x)
```

首先，先認識最基本的參數說明：  
`plt.plot()`，繪製趨勢圖：  
1. x: x軸資料
2. y: y軸資料
3. ls: 線條格式
4. lw: 線條寬度
5. label: 這圖的標記(需搭配`plt.legend`才會顯示)


`plt.legend(loc=)`，顯示label，利用參數`loc`設置位置  
`plt.show()`，顯示圖表，但是在jupyter中如果利用`%matplotlib inline`設置就可以在不執行該指令也可以顯示圖表

```python
plt.plot(x, y, ls='--', lw=2, label='sample plot')
plt.legend(loc='best')
plt.show()
```

動手做做看：  
1. 試著調整`y=cos(x)`可以觀察到label的位置變化了，這是因為我們設置`loc=best`
2. 設置`ls='-'`，觀察線條變化
3. 設置`lw=5`，觀察線條變化

趨勢圖可以觀察資料的走勢，這在訓練神經網路之後觀察收斂情況或是股票走勢的時候非常實用。
