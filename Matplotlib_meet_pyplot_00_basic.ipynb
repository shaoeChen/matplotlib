載入需求套件

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

使用`matplotlib`之前，我們先簡單理解圖表的組成。

圖表本身是一個`figure`，一個`figure`可以包含多個`axes`，各`axes`可以再自定義自己的顯示內容。

這邊我們只是要先認識圖表的組成，因此先不會談到多個`axes`或多個`figure`的應用，先理解圖表上的各組件吧。

在開始繪製圖表之前，我們先利用`numpy`來產生一些測試資料

* `np.linspace(start, stop, num)`，在start與stop之間產生num筆資料
* `np.random.randint(low, high, size)`，產生low到high之間size的整數亂數數值

```python
x = np.linspace(1, 10, 100)
y = np.random.randint(1, 10, 100)
```

下面是一個標準圖表，給定資料x、y，利用`plt.scatter`來產生散佈圖
1. x: x軸資料
2. y: y軸資料

```python
plt.scatter(x, y)
plt.show()
```

如果一個axes上有多個資料集的話，我們需要`label`來標記，一眼望去就知道這是那一個資料，這需要搭配`plt.legend()`

* `plt.legend(loc='best')`，顯示圖表中所設置的`label`，參數`loc`可以指定位置，設置上也非常直觀`loc='lower left'`就是指左下。

```python
plt.scatter(x, y, label='scatter sample')
plt.legend(loc='best')
plt.show()
```

圖表上需要設置軸的範圍這可以利用`plt.xlim`、`plt.ylim`來實現，並且需要知道x、y軸各自代表的意義，這部份可以利用`plt.xlabel`、`plt.ylabel`來實現

```python
plt.scatter(x, y, label='scatter sample')
plt.legend(loc='best')
# 故意將值軸的最大、最小值比數據資料更小，讓圖表有留白的空間
plt.xlim(min(x)-1, max(x)+1)
plt.ylim(min(y)-1, max(y)+1)
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.show()
```

圖表輔助線：
* 網格: plt.grid()
* 水平線: plt.axhline(y, c, ls, lw)
* 垂直線: plt.axvline(x, c, ls, lw)

基本上，函數內的參數都是通用的，如下說明：
* c: 顏色
* ls: 線條型式，但只有線條類的有，如果散佈點用了會拋出異常
* lw: 線條(點)大小

```python
plt.scatter(x, y, label='scatter sample')
plt.legend(loc='best')
# 設置網格
plt.grid(ls='--')
# 設置水平線
plt.axhline(y=2, c='red', ls='--', lw=2)
# 設置垂直線
plt.axvline(x=2, c='green', ls='-', lw=3)
plt.show()
```

上面範例故意使用不同的顏色來做成區隔，`c`的參數有很多，建議直接至官方網看官方文件是最洽當的。

部份時候我們需要的是一個參考區域，而不是參考線，這時候可以改以`plt.axvspan`來實現  
* 垂直參考區域: `plt.axvspan(xmin, xmax, facecolor,  alpha)`
* 水平參考區域: `plt.axhspan(ymin, ymax, facecolor,  alpha)`

參數說明如下：
* min: 最小
* min: 最大
* facecolor: 區域色彩
* alpha: 透明度

```python
plt.scatter(x, y)
plt.axvspan(xmin=2, xmax=5, facecolor='r',  alpha=0.5)
plt.show()
```

```python
plt.scatter(x, y)
plt.axhspan(ymin=2, ymax=5, facecolor='yellow',  alpha=0.5)
plt.show()
```

這邊案例上，參數`c`所使用方式不同，因此這部份的需求務必至官方文件確認。

最後，圖表需要標題，很直觀的就是利用`plt.title`來實作就可以

```python
plt.scatter(x, y)
plt.title('Scatter')
plt.show()
```
