```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

`matplotlib.rcParams`是`matplotlib`配置文件的操作，詳情可見[官方連結](https://matplotlib.org/api/matplotlib_configuration_api.html)

下面利用`find_all`來看所有可用的參數

```python
mpl.rcParams.find_all
```

如果你有中文顯示的需求，那就需要調整配置才能達到

```python
x = [1, 2, 3]
y = [1, 2, 3]
```

```python
plt.plot(x, y)
plt.title('中文')
plt.show()
```

上面可以看的到，中文兩個字在預設情況下是無法顯示的，但要特別注意自己有沒有相關的字型。

```python
mpl.rcParams['axes.unicode_minus']=False
mpl.rcParams['font.sans-serif']=['LiSu']
```

```python
plt.plot(x, y)
plt.title('中文')
plt.show()
```
