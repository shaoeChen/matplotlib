## 載入需求套件  
下面三個套件是在python上不可或缺的好幫手。  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  語法糖，不用每次都plt.show()來呈現圖表
%matplotlib inline
```

設置簡單資料集

```python
#  簡單就好
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [8, 7, 6, 5, 4, 3, 2, 1]
```

## 加入子圖表
如果要像股票一樣，有個大表可能是個股，然後小表是目前的股市表現!

```python
#  定義一個大相框
fig = plt.figure(figsize=(8, 4))
#  你要的個股資料
#  左下(0.1, 0.1)與右上(0.8, 0.8)的位置
#  如果設置1就全開紙了
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#  加入股市表現
ax = fig.add_axes([0.15, 0.15, 0.3, 0.3])
plt.show()
```

## 多圖表  
不只是在加入子圖表的時候使用，一般情況下也是可以這樣子使用，現在在網路看到的大部份範例(含原廠文件)也是這樣子用!  
使用subplots的時候，會回傳兩個物件：    
1. fig(figure):把它想成是一個有N格相框的『大相框』!
2. ax(axes或稱軸):把它想成是上面N格相框中的其中一個『相框』! 

後續的執行動作都是一樣的。

```python
fig, ax = plt.subplots(figsize=(8, 4),  #  大相框尺寸
                       dpi=100,  #  每英吋幾個點
                       tight_layout=True,  #  把所有東西限縮在大相框內
                       linewidth=1,  #  寬
                       edgecolor='r')  #  大相框顏色
```

```python
fig, ax = plt.subplots(figsize=(8, 4),  #  大相框尺寸
                       dpi=100,  #  每英吋幾個點
#                        tight_layout=True,  #  把所有東西限縮在大相框內
                       linewidth=1,  #  寬
                       edgecolor='r')  #  大相框顏色
scatter = ax.scatter(x, y, label='TestScatter')
ax.legend(loc='best')
ax.set_title('TestScatter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
fig.text(0.995,0.01,'test',ha='right',va='bottom')
fig.tight_layout()
plt.show()
```

透過`nrows`與`ncols`來設置子圖表數(ax)  
下面範例我們設置了(3, 1)的圖表樣式

```python
fig, ax = plt.subplots(nrows=3,  #  要三個橫軸
                       ncols=1,  #  要一個直軸
                       sharex=True,  #  是否共用x軸座標
                       sharey=True,  #  是否共用y軸座標
                       figsize=(8,4))
ax0 = ax[0]
ax0.scatter(x,y, label='axes0')
ax0.legend(loc='best')
ax1 = ax[1]
ax1.scatter(x,y, label='axes1')
ax1.legend(loc='best')
ax2 = ax[2]
ax2.scatter(x,y, label='axes2')
ax2.legend(loc='best')
plt.tight_layout()
plt.show()
```

除了透過`fig, ax = plt.subplots`的設置方式之外，也可以利用`plt.subplot`來設置  
當你執行了`plt.subplot`時，此時的圖表控制權就是在這個軸上。

```python
plt.subplot(221)
plt.scatter(x, y, label='221')
plt.legend(loc='best')
plt.subplot(222) 
plt.scatter(x, y, label='222')
plt.legend(loc='best')
plt.subplot(212) 
plt.scatter(x, y, label='211')
plt.legend(loc='best')
plt.show()
```
