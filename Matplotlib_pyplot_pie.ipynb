載入需求套件

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

建立資料

```python
pie_labels = ['A', 'B', 'C', 'D', 'E', 'F']
pie_rate = [0.3, 0.1, 0.2, 0.1, 0.2, 0.1]
```

首先，先認識最基本的參數說明：  
`plt.pie()`，繪製園餅圖：  
1. x: 值
2. labels: 各值域的標籤
3. autopct: 格式化字串
4. colors: 顏色

`plt.show()`，顯示圖表，但是在jupyter中如果利用`%matplotlib inline`設置就可以在不執行該指令也可以顯示圖表

```python
plt.pie(pie_rate,
        labels=pie_labels,
        autopct='%3.1f%%',)
plt.title('Pie Chart')
plt.show()
```

動手做做看：  
1. 試著加入參數`startangle=30`來觀察圖形的變化
2. 試著加入參數`colors=[]`來控制顏色

圓餅圖可以直觀的瞭解各類別的佔比，也是經典實用圖之一。
